```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'rollno': [101, 102, 103, 104, 105, 106],
    'name': ['Anil','Bhumika', 'Chetan', 'Dhruv', 'Emaanuel', 'Farah'],
    'marks': [78, 93, 85, 96, 12, 82],
    'grade': ['C','A', 'B', 'A', 'F', 'B'],
    'city': ['Pune', 'Mumbai', 'Mumbai', 'Delhi', 'Delhi', 'Delhi']
},
index=[1,2,3,4,5,6]
)

df
```

```python
# group by city
df.groupby('city').agg({
    'city': ['count']
})
```

```python
# select students whose marks greater than 80
df[df['marks'] > 80]
```

```python
# select students where marks greater than 80 and city is mumbai

df[(df['marks'] > 80) & (df['city'] == 'Mumbai')]
```

```python
# select students where marks greater than 80 or city is mumbai

df[(df['marks'] > 80) | (df['city'] == 'Mumbai')]
```

```python
# select students whose makrs lies in between 80 and 90

df[(df['marks'] >= 80) & (df['marks'] <= 90)]
```

```python
# selects students of 'Delhi' and 'Mumbai'

df[df['city'].isin(['Delhi', 'Mumbai'])]
```
