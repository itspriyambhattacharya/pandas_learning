# Pandas Data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\coding\pandas\datasets\Ice Cream Ratings.csv")
df
```

```python
df = df.set_index('Date')
df
```

```python
df.plot()
```

```python
df.plot(kind='line', title='Ice Cream Report', xlabel='Daily Rating', ylabel='Scores')
```

```python
df.plot(kind='bar', title='Ice Cream Report Bar Graph', xlabel='Daily Report', ylabel='Scores')
```

```python
df.plot(kind='bar', stacked=True, xlabel='Daily Rating', ylabel='Scores', title='Ice Creame Stacked Bar Graph Report')
```

```python
df['Flavor Rating'].plot(title='Flavor Rating Plot', xlabel='Flavor Data', ylabel='Score', kind='bar')
```

```python
df.plot.barh(stacked='True')
```
