# Difference between loc and iloc

```python
df2 = pd.DataFrame({
    "A": [1,2,3],
    "B": [4,5,6],
    "C": [7,8,9],
    "D": [10, 11, 12]
}, index=["i1", "i2", "i3"])

print(df2)
print(f"Value of index i1 and column A using loc is: {df2.loc["i1", "A"]}")
print(f"Value of index i1 and column C using loc is: {df2.loc["i1", "C"]}")
print(f"Value of index i1 to i2 and column D using loc is: {df2.loc["i1":"i2", "D"]}")
print(f"Value of index i1 to i3 and column D using loc is: {df2.loc["i1":"i3", "A"]}")
print(f"Value of index i1 to i3 and column ALL using loc is: {df2.loc["i1":"i3", ]}")
print(f"Value of index i1 to i3 and column A and B using loc is: {df2.loc["i1":"i3", "A":"B"]}")
```

```python
df3 = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6],
    "C": [7, 8, 9],
    "D": [10, 11, 12]
}, index=["i1", "i2", "i3"])

print(df3)

print(f"Value of index 1 and column 1 using iloc: {df3.iloc[1,1]}")
print(f"Value of first 2 indices of first 3 columns using iloc: {df3.iloc[0:2,:3]}")
```
