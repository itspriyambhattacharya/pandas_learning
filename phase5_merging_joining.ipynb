# Phase 5 - Merging, Joining in Pandas

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\coding\pandas\datasets\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"D:\coding\pandas\datasets\LOTR 2.csv")
df2
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName']) # inner join
```

```python
df1.merge(df2, how='outer') # full outer join
```

```python
df1.merge(df2, how='left') # left outer join
```

```python
df1.merge(df2, how='right', on=['FellowshipID', 'FirstName']) # right-outer join
```

```python
df1.merge(df2, how='cross') # cross-product
```

```python
pd.concat([df1, df2], axis = 1)
```
